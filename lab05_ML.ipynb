```python
# Extracting athe 4th and 5th coloumn from dataset 
data4=data.iloc[:,4]
data5=data.iloc[:,5]
r_values = np.arange(1, 11)

# Calculate minkowski distance
from scipy.spatial import distance
distances = [distance.minkowski(data4,data5, p=r) for r in r_values]

plt.plot(r_values, distances, marker='o')
plt.title('Minkowski Distance vs. r')
plt.xlabel('r')
plt.ylabel('Minkowski Distance')
plt.grid(True)
plt.show()
```

```python
# Extracting features and target from the dataset 
features = data.iloc[:, 0:383]
target = data.iloc[:,384]
target = target.astype(int)
print(target)
print(features)
# Splitting the data into test and train data
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.3,random_state=42)
```

```python
## Train a kNN classifier with k=3
knn_classifier = KNeighborsClassifier(n_neighbors=3)
knn_classifier.fit(X_train, y_train)
y_pred = knn_classifier.predict(X_test)
y_train_pred = knn_classifier.predict(X_train)
y_test_pred = knn_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Train a kNN classifier with k=1
one = KNeighborsClassifier(n_neighbors=1)
one.fit(X_train, y_train)
y_pred_one = one.predict(X_test)
accuracy_one = accuracy_score(y_test, y_pred_one)

# Train a kNN classifier with k=3
three = KNeighborsClassifier(n_neighbors=3)
three.fit(X_train, y_train)
y_pred_three = three.predict(X_test)
accuracy_knn_3 = accuracy_score(y_test, y_pred_three)

# Evaluate accuracy for different values of k
k_values = range(1, 12)
accuracy_scores_knn = []

for k in k_values:
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(X_train, y_train)
    y_pred_knn = knn_classifier.predict(X_test)
    accuracy_knn = accuracy_score(y_test, y_pred_knn)
    accuracy_scores_knn.append(accuracy_knn)

# Plot the accuracy vs. k for kNN
plt.figure(figsize=(10, 6))
plt.plot(k_values, accuracy_scores_knn, marker='o', linestyle='-', color='b', label='kNN (k varying)')
plt.axhline(y=accuracy_one, linestyle='--', color='r', label='NN (k=1)')
plt.axhline(y=accuracy_knn_3, linestyle='--', color='g', label='kNN (k=3)')
plt.title('Accuracy vs. k for kNN')
plt.xlabel('k')
plt.ylabel('Accuracy')
plt.legend()
plt.show()



```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import numpy as np

# Load the data
sheet_name = "IRCTC Stock Price" 
data2 = pd.read_excel(r"C:\Users\91630\OneDrive\Desktop\ml\ml lab4\Lab Session1 Data.xlsx", sheet_name=sheet_name)
print(data2)

# Convert date to numeric using label encoding
label_encoder = LabelEncoder()


# Convert categorical values to numeric using label encoding
categorical_columns = ['Date', 'Month',"Day","Volume"]  # Replace with your actual column names
for col in categorical_columns:
    data2[col] = label_encoder.fit_transform(data2[col])

# Separate features and target
features = data2.drop('Price', axis=1)  
target = data2['Price']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Train the linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Add predicted values to the original dataframe
data2['predicted_column'] = np.nan 
data2.loc[X_test.index, 'predicted_column'] = y_pred

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("Mean Absolute Error:", mae)
print("R2 Score:", r2)
```

```python
np.random.seed(42)
# Generate random training data
X_train = np.random.uniform(1, 10, 20)
Y_train = np.random.uniform(1, 10, 20)
# Create binary classes based on the condition (X_train + Y_train > mean)
class_train = np.where((X_train + Y_train) > np.mean(X_train + Y_train), 1, 0)

# Initialize k-NN model  with 3 neighbors
knnc = KNeighborsClassifier(n_neighbors=3)

# Fit the classifier on the training data
knnc.fit(np.column_stack((X_train, Y_train)), class_train)

# Generate a grid of test points
X_test = np.arange(0, 10.1, 0.1)
Y_test = np.arange(0, 10.1, 0.1)
test_set = np.array(np.meshgrid(X_test, Y_test)).T.reshape(-1, 2)

# Predict classes for the test set
predicted_classes = knnc.predict(test_set)

# Plot the test data points with predicted classes
plt.scatter(test_set[predicted_classes == 0, 0], test_set[predicted_classes == 0, 1], color='green', alpha=0.5, s=5, label='0')
plt.scatter(test_set[predicted_classes == 1, 0], test_set[predicted_classes == 1, 1], color='black', alpha=0.5, s=5, label='1')

# Plot the training data points with their actual classes
plt.scatter(X_train[class_train == 0], Y_train[class_train == 0], color='purple', marker='o', s=50, label='Train 0')
plt.scatter(X_train[class_train == 1], Y_train[class_train == 1], color='red', marker='o', s=50, label='Train 1')

# Set labels for the plot and dispaly the plot 
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Plot of Test Data with kNN Classifier')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsRegressor
# feature1, feature2, and classes are the columns from the dataset
feature1 = data1
feature2 = data3
target = classes

# Ensure all arrays are of the same length
min_length = min(len(feature1), len(feature2), len(target))
feature1 = feature1[:min_length]
feature2 = feature2[:min_length]
target = target[:min_length]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(np.column_stack((feature1, feature2)), target, test_size=0.2, random_state=42)

# Use KNeighborsRegressor for regression
knn_regressor = KNeighborsRegressor(n_neighbors=3)
knn_regressor.fit(X_train, y_train)

# Create a consistent range for both features
feature1_range = np.arange(feature1.min(), feature1.max() + 0.1, 0.1)
feature2_range = np.arange(feature2.min(), feature2.max() + 0.1, 0.1)

# Ensure both ranges have the same length
min_range_length = min(len(feature1_range), len(feature2_range))
feature1_range = feature1_range[:min_range_length]
feature2_range = feature2_range[:min_range_length]

# Create the test_set DataFrame with consistent ranges
test_set = pd.DataFrame({
    'Feature1': feature1_range,
    'Feature2': feature2_range
})

# Predict values for the test set
predicted_values = knn_regressor.predict(test_set[['Feature1', 'Feature2']])

# Scatter plot of test data colored as per predicted values
plt.scatter(test_set['Feature1'], test_set['Feature2'], c=predicted_values, cmap='viridis', alpha=0.5, s=5, label='Predicted Values')

# Plot training data points
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap='viridis', marker='o', s=50, label='Training Data')

plt.xlabel('Feature1')
plt.ylabel('Feature2')
plt.title('Scatter Plot of Test Data with kNN Regression')
plt.legend()

plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameters to search
param_grid = {'n_neighbors': range(1, 21)}  

# Create the GridSearchCV object
grid_search = GridSearchCV(knn_regressor, param_grid, cv=5, scoring='neg_mean_squared_error')
# Fit the model to the training data
grid_search.fit(X_train, y_train)
# Print the best parameters and corresponding negative mean squared error
print("Best Parameters: ", grid_search.best_params_)
print("Best Negative Mean Squared Error: ", grid_search.best_score_)
```
